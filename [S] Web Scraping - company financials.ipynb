```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = ['Revenue_2021','Revenue_2020','Revenue_2019','EBITDA_2021','EBITDA_2020','EBITDA_2019']
current_company_financials = pd.DataFrame(columns=headers)


ticker = ['MSFT','NOW','TTD','AAPL', 'DE','CRM']
ticker_index = 0
num = 0

for ticker_index in range(len(ticker)):
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.67 Safari/537.36"}
    url = f'https://stockanalysis.com/stocks/{ticker[ticker_index]}/financials/'
    page = requests.get(url, headers=headers)
    soup = BeautifulSoup(page.text, 'html.parser')
    raw_data = soup.find_all('tr')
    
    financial_data = []
    financials_index = [1,  -4]  #extract Revenue, EBITDA
    financials_year_index = [ 1, 2, 3]    #extract title, data of 2021, 2020 & 2019

    for i in financials_index:
        for j in financials_year_index:
            cell_data = raw_data[i].find_all('span')[j].string
            financial_data.append(cell_data)
    
    
   
    current_company_financials.loc[num] = financial_data
    #current_company_financials.index = ticker[ticker_index]
    num = num + 1
    #ticker_index = ticker_index + 1

current_company_financials.index = ticker
current_company_financials
```
